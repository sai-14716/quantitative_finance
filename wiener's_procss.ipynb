```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = [0]
nsteps = 100
tstep = 1

for i in range(nsteps):
    w.append(np.random.normal(0,tstep,1)[0])
```

```python
w1 = [0]

for i in range(nsteps):
    sum = 0
    for j in range(i):
        sum += np.random.normal(0,i-j,1)[0]
    w1.append(sum/(i+1))
```

```python
plt.xlabel("Time")
plt.ylabel("Value using normal")
plt.plot(w)
plt.show()

plt.plot(w1)
plt.show()
```

```python
from scipy.interpolate import interp1d

x = np.linspace(0,100,101)

a = interp1d(x, w, kind='linear')

y = a(x)

plt.plot(x,y)
```
