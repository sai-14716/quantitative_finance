The geometric brownian motion is the commonly used model for the stock pricing 
dX(t) = mu * X(t) * dt + (volatility) * (X(t)) * dW(t)

Using ito's lemma it gets to
dX(t) = (mu - 1/2 *volatility ** 2) * dt + (volatility) * dW(t)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the initial conditions
T = 1
mu = 0.2
volatility = 0.4
no_of_steps = 200
dt = T/no_of_steps
no_of_paths = 100


#Simulating the paths
t = 0
X = np.zeros([no_of_paths, no_of_steps])
W = np.random.normal(0, 1, [no_of_paths,no_of_steps])
timegrid = np.zeros([no_of_steps])

for i in range(1, no_of_steps):
    t = t + dt
    X[:, i] = X[:, i-1] + (mu - 0.5 * (volatility ** 2)) * dt + (volatility * np.random.normal(0, t ** 0.5, no_of_paths))
    timegrid[i] = t

S = np.exp(X)
mean = np.mean(X)
print(mean)

plt.plot(timegrid,S.T)
plt.xlabel("Time")
plt.grid()
plt.ylim(0,200)
plt.ylabel("STock price")
plt.title("GBM")
plt.show()
```
