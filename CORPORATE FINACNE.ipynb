```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
```

### PLOTTING THE RETURNS VS RISK/SD/VARIANCE FOR 2 SECURITIES A,B 

```python
A = [14, 0.2]
B = [10, 0]
```

```python
x = np.linspace(0,100,5)
y = B[0] + (A[0]-B[0]) * (x/100) #returns
z = A[1] * x
```

```python
ax = plt.axes(projection = '3d')
ax.grid()

ax.plot3D(x, y, z)
plt.show()
```
