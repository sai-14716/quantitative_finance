## Predicting the Black scholes solution by Martingales and Analytical Solution 

```python
import numpy as np
import pandas as pd
import scipy as sp
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#Pre-defined methods
def optimal_gbm_paths(X_orig, mu = 0, sigma = 1, start = 0, interval = 1, end = 100, paths = 1, return_optimal = 1):
    x_axis = np.arange(start, end, interval) 
    times = int((end - start)/interval)
    X = np.full([times ,paths], np.log(X_orig[0]))
    optimal = np.zeros(paths)
    # print("New Path")
    for i in range(times - 1):
        X[i + 1, :] = X[i, :] + (mu - 0.5 * (sigma ** 2)) * interval + sigma * np.random.normal(loc=0, scale=interval ** 0.5, size=paths)
        optimal = optimal + (X[i + 1, :] - np.log(X_orig[i])) ** 2
    X = np.exp(X)
    X[-1] = X_orig[-1]
    i = np.argsort(optimal)

    # print(X)
    plt.figure(1)
    for j in range(paths):
        plt.plot(x_axis, X[:, j])
    # plt.savefig('real_mu_prediction.png')

    plt.figure(2)
    for a in range(0, return_optimal):
        plt.plot(x_axis, X[:, i[a]])
        # print(optimal[i[a]])
    # plt.savefig('real_mu_prdictions_optimal(5).png')
    plt.ylim(150, 240)
    
    return X

# X_orig = np.zeros(100)
# optimal_gbm_paths(X_orig, mu = 0.019, sigma = 0.14, start = 0, interval = 1, end = 100, paths = 10, return_optimal = 5)
```

### Problem-1 : Finding the mu, sigma of the stock price
Two different solutions are using 

$\texttt{Performance measure : The squared difference error of the generated path and the original path}$
$\texttt{Method-1 : Use the mean as mu and the greatest difference as the sigma}$
$\texttt{Method-2 : (Difficult but accurate) Finnd the up and downtrends and assign the different mu's}$
$\texttt{Method-3 : (Difficult but accurate) We Will use the ML models}$

Note: In method-2 overfitting should be taken care of very carefully

```python
#Method - 1
def simple_mu_sigma(prices):
    mu = (prices[-1] - prices[0])/len(prices) 
    sigma = 0.0
    for i in range(1, len(prices)):
        sigma = max(sigma, prices[i] - prices[i - 1])
    mu += 0.5 * (sigma)
    sigma = sigma ** 0.5
    return mu, sigma

#Method - 2 (The Formalistic)
def real_mu_sigma(price):
    
    curr_trend = 'inc' if(price[1] > price[0]) else 'dec'
    previous_threshold = 0
    temporary_braking = None
    breaks = [[previous_threshold, curr_trend]]
    threshold = 0.4

    
    for i in range(len(price) - 1):
        prev = price[i]
        curr = price[i + 1]

        if curr_trend == 'dec':
            if temporary_braking is None and (curr > prev):
                temporary_braking = i

            if temporary_braking is not None: 
                if(curr < price[temporary_braking]):
                    temporary_braking = None
                elif ((curr - price[temporary_braking]) > (threshold * (price[previous_threshold] - price[temporary_braking]))):
                    # Found a new break at temporary break
                    # print(f"Foud a break at {temporary_braking}")
                    previous_threshold = temporary_braking
                    curr_trend = 'inc'
                    breaks.append([previous_threshold, curr_trend])
                    temporary_braking = None
                    
        if curr_trend == 'inc':
            if temporary_braking is None and (curr < prev):
                temporary_braking = i
            if temporary_braking is not None:
                if(curr > price[temporary_braking]):
                    temporary_braking = None
                elif ((price[temporary_braking] - curr) > (threshold * (price[temporary_braking] - price[previous_threshold]))):
                    # FOund a new break
                    # print(f"Foud a break at {temporary_braking}")
                    previous_threshold = temporary_braking
                    curr_trend = 'dec'
                    breaks.append([previous_threshold, curr_trend])
                    temporary_braking = None

    greeks = []
    for i in range(len(breaks) - 1):
        prev = breaks[i][0]
        curr = breaks[i + 1][0]
        mu_sig = (np.log(price[curr]) - np.log(price[prev]))/(curr - prev)
        sigma = 0
        for j in range(prev, curr):
            t1 = np.log(price[j + 1]) - np.log(price[j]) - mu_sig
            sigma = max(sigma,  abs(t1))
        mu = mu_sig + 0.5 * (sigma ** 2)
        greeks.append([mu, sigma])
    # print(greeks)  
    return greeks, breaks

# m, sigm = simple_mu_sigma(data) #--- The output is downlaoded in the figure named simple_predictions.png 


greeks, breaks = real_mu_sigma(data)
# print(breaks)
prices = [data[brea[0]] for brea in breaks]
x = [brea[0] for brea in breaks]
plt.figure(3)
plt.plot(x, prices)
plt.plot(data)

# print(greeks)
# print(m, sigm)
# optimal_gbm_paths(data, mu= m /2000, sigma= sigm /200, start = 0, interval = 1, end = 100, paths = 100, return_optimal = 5)
for i in range(1, len(greeks)):
    start = breaks[i - 1][0]
    end = breaks[i][0]
    optimal_gbm_paths(data[start:end + 1], mu = -greeks[i][0], sigma = -greeks[i][1], start = start, end = end, paths = 10, return_optimal = 1)
```

## Option Pricing using the Analytical Solution
$\texttt{We Will price the option using the Black Scholes Merton Analytical Solution}$
$$ V_c(t, S(t)) = S(t)N(d_1(t, S(t))) - e^{-r(T-t)}KN(d_2(t, S(t)) $$
$$ d_1(t, S(t)) = \frac{1}{\sigma \sqrt{T-t}}(log\frac{S(t)}{K} + (r + \frac{\sigma ^ 2}{2})(T - t)) $$
$$ d_2(t, S(t)) = d_1(t, S(t)) - \sigma \sqrt{T-t} $$

### Mathematically facts about the analytical solution
The option price is a monotonic function of $\sigma$ : Because vega:V is always > 0 

```python
from scipy.stats import norm

class black_scholes:
    def __init__(self, mu, sigma, r):
        self.mu = mu
        self.sigma = sigma
        self.rate = r
        
    def analytical_option_price(self, s, t, k, T):
        d_1 = self._find_d1(s, t, k, T)
        d_2 = self._find_d2(s, t, d_1, T)
        v = s * norm.cdf(d_1) - np.exp(-self.rate * (T - t)) * k * norm.cdf(d_2)
        return v
    
    def _find_d1(self, s, t, k, T):
        return (np.log(s/k) + (self.rate + (self.sigma ** 2)/2) * (T - t))/(self.sigma * (T - t)**0.5)
    
    def _find_d2(self, s, t, d_1, T):
        return d_1 - (self.sigma * (T - t)**0.5)

    def martingale_option_price(self, s, t, k, T):
        x = np.full([T - t], s)
        X = optimal_gbm_paths(x, start = t, end = T, paths = 100)
        print(np.mean(X[-2, :]))
        
#Examples
a = black_scholes(0.007, 0.025, 0.05)
print(a.analytical_option_price(100, 0, 100, 1))
print(a.analytical_option_price(100, 0, 100, 20))
a.martingale_option_price(100, 0, 100, 20)
```
