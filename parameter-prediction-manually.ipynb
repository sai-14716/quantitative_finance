## Prediction $\mu$ and $\sigma$ using the common closed for maths equation and $\sigma$ as max($t_{(i-1)}$ ~ $t_{i})$ 

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
stock = 'AAPL'
print(yf.__version__)
data = yf.download(stock, period = '3mo', interval = '1h', auto_adjust=True)

data = data.to_numpy()
data = data[:, 1].reshape(len(data), 1)
x = np.linspace(0, len(data), len(data))
print(data.shape, x.shape, len(data))
from sklearn.preprocessing import StandardScaler
data = StandardScaler().fit_transform(data)
plt.plot(data)
plt.scatter(x, data)
#finding the sigma
max = 0
y = data[0]
x = 0
for i in range(1, len(data)):
    diff = data[i] - data[i - 1]
    if abs(diff) > max:
        max = abs(diff)
    y += data[i]
    x += i
print(max, y/x)

```
