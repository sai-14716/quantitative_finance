```python
def f(x):
    return (x[0]-1) ** 2 + (x[1] - 2) ** 3

sp.optimize.minimize(f, (2.5,3.10))
```

```python
from scipy.stats import beta, norm

a , b = 9.52, 4.75

c = beta.stats(a, b)
d = norm.stats(a, b)
print(c,d)

norm.rvs(a, b, 10)
```

```python
f = np.linspace(beta.ppf(0.10, a, b),beta.ppf(0.89, a ,b))
g = beta.pdf(f, a, b)

print(f.shape,g.shape)

plt.plot(f, g)
```
